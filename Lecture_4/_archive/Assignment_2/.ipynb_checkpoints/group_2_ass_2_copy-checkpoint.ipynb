```python
import savReaderWriter as sav
all_data = {}

years = [2019, 2018, 2017, 2016, 2015]#
for year in years:

  
    d1 = pd.read_spss( fr"../../_data/endes/{year}/REC0111.sav" )
    with sav.SavHeaderReader( fr"../../_data/endes/{year}/REC0111.sav", ioUtf8=True) as header:
        metadata = header.all()
        var_labels1 = metadata.varLabels.copy()
        value_labels1 = metadata.valueLabels.copy()
    

    
    new_var_labels1= {}
    new_value_labels1 = {}
    
    rec_1 = d1[var1]

    for key in var1:

        if key in var_labels1: # Store if select var label exist
            new_var_labels1[f'{key}'] = var_labels1[f'{key}']
        if key in value_labels1:
            new_value_labels1[f'{key}'] = value_labels1[f'{key}']
            
    #DATA 2
    

    d2 = pd.read_spss( fr"../../_data/endes/{year}/RE223132.sav" )
    with sav.SavHeaderReader( fr"../../_data/endes/{year}/RE223132.sav", ioUtf8=True) as header:
        metadata = header.all()
        var_labels2 = metadata.varLabels.copy()
        value_labels2 = metadata.valueLabels.copy()
    
    rec_2 = d2[var2]
    
    new_var_labels2= {}
    new_value_labels2 = {}
    

    for key in var2:

        if key in var_labels2: # Store if select var label exist
            new_var_labels2[f'{key}'] = var_labels2[f'{key}']
        if key in value_labels1:
            new_value_labels2[f'{key}'] = value_labels2[f'{key}']
    
    
    #DATA 3
    

    d3 = pd.read_spss( fr"../../_data/endes/{year}/RE516171.sav" )
    with sav.SavHeaderReader( fr"../../_data/endes/{year}/RE516171.sav", ioUtf8=True) as header:
        metadata = header.all()
        var_labels3 = metadata.varLabels.copy()
        value_labels3 = metadata.valueLabels.copy()
    

    rec_3 = d3[var3]
    
    new_var_labels3= {}
    new_value_labels3 = {}
    

    for key in var3:

        if key in var_labels3: # Store if select var label exist
            new_var_labels3[f'{key}'] = var_labels3[f'{key}']
        if key in value_labels1:
            new_value_labels3[f'{key}'] = value_labels3[f'{key}']
            
    #MERGE ALL DATA
    merge1 = rec_1.merge(rec_2,on = 'CASEID',how = 'left',validate = "1:1")
    merge2 = merge1.merge(rec_3,on = 'CASEID',how = 'left',validate = "1:1")
                

    ## Labeling

    new_var_label = new_var_labels1.copy()
    new_value_label = new_value_labels1.copy()

    # Add label 2
    new_var_label.update( new_var_labels2 )
    new_value_label.update( new_value_labels2 )

    # Add label 3
    new_var_label.update( new_var_labels3 )
    new_value_label.update( new_value_labels3 )

    # LABEL
    var_labels = new_var_label.copy()
    value_labels = new_value_label.copy()
    
 ## Store 
    store = {"data" : merge2 , "var_labels" : var_labels, "value_labels" : value_labels}
                    
    all_data[ f'year_{year}'] = store

```

```python
from os.path import isfile, join, isdir

os.getcwd()

ruta = r"../../_data/endes/"
contenido = os.listdir(ruta)
carpetas =  [nombre for nombre in contenido if isdir(join(ruta, nombre))]
archivos  =  ['/REC0111.sav','/RE223132.sav','/RE516171.sav']

all_data = {} #nested dictionary
list_df = [] #additional variable

for c in carpetas:  
    #dictionaries-year
    dic_year = {} 
    for arch in archivos:
        path_file = ruta + c  + arch
        #print(path_file)
        if arch == '/REC0111.sav':         
            rec11, meta11 = pyreadstat.read_sav(path_file)
            value_labels11 = meta11.variable_value_labels
            var_labels11 = meta11.column_names_to_labels
        
            rec11_1=rec11.loc[ : , ["CASEID", "V000", "V001", "V002", "V003", "V004", "V008", "V009", "V010", "V011", "V012", "V024", "V102", "V120", "V121", "V122", "V123", "V124", "V125", "V127", "V133"] ]
            
            new_var_labels11 = {key: var_labels11[key] for key in rec11_1.keys()}
            new_value_labels11 = {key: value_labels11 [key] for key in value_labels11 if key in new_var_labels11.keys()}
            
            rec11_1.loc[:,'year']=c
            year = {"year" : "Año de la encuesta"}                      
            new_var_labels11.update(year)
            new_var_labels11
            #print(path_file)
        if arch == '/RE223132.sav':         
            rec22, meta22 = pyreadstat.read_sav(path_file)
            value_labels22 = meta22.variable_value_labels
            var_labels22 = meta22.column_names_to_labels
        
            rec22_1=rec22.loc[ : , ["CASEID", "V201", "V218", "V301", "V302", "V323", "V323A", "V325A", "V326", "V327", "V337", "V359", "V360", "V361", "V362", "V363", "V364", "V367", "V372", "V372A", "V375A", "V376", "V376A", "V379", "V380"] ]

            new_var_labels22 = {key: var_labels22[key] for key in rec22_1.keys()}
            new_value_labels22 = {key: value_labels22 [key] for key in value_labels22 if key in new_var_labels22.keys()}
            #print(path_file)       
        if arch == '/RE516171.sav':         
            rec33, meta33 = pyreadstat.read_sav(path_file)
            value_labels33 = meta33.variable_value_labels
            var_labels33 = meta33.column_names_to_labels
        
            rec33_1=rec33.loc[ : , ["CASEID", "V501", "V502", "V503", "V504", "V505", "V506", "V507", "V508", "V509", "V510", "V511", "V512", "V513", "V525", "V613", "V714", "V715"] ]

            new_var_labels33 = {key: var_labels33[key] for key in rec33_1.keys()}
            new_value_labels33 = {key: value_labels33 [key] for key in value_labels33 if key in new_var_labels33.keys()}
            #print(path_file)           
    
    data_c = rec11_1.merge( rec22_1 , on = [ 'CASEID' ] , how = "left" , validate = "m:1" )
    data_c = data_c.merge( rec33_1 , on = [ 'CASEID' ] , how = "left" , validate = "m:1" )
        
    new_var_labels11.update(new_var_labels22)
    new_var_labels11.update(new_var_labels33)
    var_labels = new_var_labels11.copy()

    new_value_labels11.update(new_value_labels22)
    new_value_labels11.update(new_value_labels33)
    value_labels = new_value_labels11.copy()

    data_c.attrs[ 'var_labels' ] = var_labels
    data_c.attrs[ 'value_labels' ] = value_labels
    list_df.append(data_c)
    
    #dictionary is created - year
    dic_year[f'data']= data_c
    dic_year[f'var_labels'] = var_labels
    dic_year[f'value_labels' ] = value_labels
    
    #print(c)
    
    #nests dictionaries -all
    all_data[f'year_{c}']= dic_year
    print(c)
    
#df = pd.concat(list_df, ignore_index = True)
print(all_data) 
```

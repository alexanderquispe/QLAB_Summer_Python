```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver import ActionChains
```

```python
import pandas as pd
import numpy as np
import os
import time
import re
import unidecode
```

```python
from selenium.webdriver.support.ui import Select
from webdriver_manager.chrome import ChromeDriverManager
```

```python
driver = webdriver.Chrome( ChromeDriverManager().install() )

url = 'https://resultadoshistorico.onpe.gob.pe/EG2021/EleccionesPresidenciales/RePres/E'
driver.get( url )

driver.maximize_window()
```

```python
# Store all_tables
all_tables = {}

# All selenium objects in 
cont = Select( driver.find_element_by_xpath( "/html/body/onpe-root/onpe-layout-container/onpe-onpe-epres-re/div[1]/div[3]/div[1]/div[1]/div/div/div[2]/select" ) )

# Get number of total options
num_cont_options = len( cont.options )
```

```python
num_cont_options
```

```python
for cont_idx in range( num_cont_options ):

    # Click on one specific continent:
    cont = Select( driver.find_element_by_xpath( "/html/body/onpe-root/onpe-layout-container/onpe-onpe-epres-re/div[1]/div[3]/div[1]/div[1]/div/div/div[2]/select" ) )
    continent = cont.options[ cont_idx ]
    
    # Get continent name:
    cont_name = continent.text
    
    # We select a different continent name
    if cont_name != "--TODOS--" :
        
        # click on continent
        continent.click()
        
        # Get all elements of country
        country = Select( driver.find_element_by_xpath( "/html/body/onpe-root/onpe-layout-container/onpe-onpe-epres-re/div[1]/div[3]/div[1]/div[1]/div/div/div[3]/select" ) )
        num_country_options = len( country.options )
        
        for country_idx in range( num_country_options ):
            
            # Get again all countries since HTML is refreshing
            country = Select( driver.find_element_by_xpath( "/html/body/onpe-root/onpe-layout-container/onpe-onpe-epres-re/div[1]/div[3]/div[1]/div[1]/div/div/div[3]/select" ) )
            countries = country.options[ country_idx]
                
            # Get country name
            country_name = countries.text
            
            if country_name != "--TODOS--" :
                
                # click on country
                countries.click()
                
                # Get all elements from state
                state = Select( driver.find_element_by_xpath( "/html/body/onpe-root/onpe-layout-container/onpe-onpe-epres-re/div[1]/div[3]/div[1]/div[1]/div/div/div[4]/select" ) )
                num_state_options = len( state.options )
                
                for state_idx in range( num_state_options ):
                    
                    # Get again all states since HTML is refreshing
                    state = Select( driver.find_element_by_xpath( "/html/body/onpe-root/onpe-layout-container/onpe-onpe-epres-re/div[1]/div[3]/div[1]/div[1]/div/div/div[4]/select" ) )
                    states =  state.options[ state_idx ]
                    
                    # Get state name
                    state_name = states.text
                    
                    if state_name != "--TODOS--" :
                        
                        # click on state
                        states.click()
                        
                        # Get UBIGEO and link using URL:
                        ubigeo = driver.current_url.split("/")[ -1 ]
                        
                        link = driver.current_url
                        
                        ## Get table of presidential votes
                        table_path = driver.find_element_by_xpath( "/html/body/onpe-root/onpe-layout-container/onpe-onpe-epres-re/div[1]/div[4]/div[1]/div[4]" )
                        table_html = table_path.get_attribute( 'innerHTML' )
                        # Read the table using pandas
                        table = pd.read_html( table_html )
                        
                        # Cleaning tables
                        row_new_columns = table[ 0 ].iloc[ 0 , 2: ]
                       
                        # Selecting specific columns
                        table_clean = table[0].iloc[ 1:, 2: ].copy()
                                             
                        # New values to columns 
                        table_clean[ 'continent' ] = cont_name
                        table_clean[ 'country' ]   = country_name
                        table_clean[ 'state' ]   = state_name
                        table_clean[ 'ubigeo' ]     = ubigeo
                        table_clean[ 'link' ]     = link
                        
                        # store tables
                        all_tables[ ubigeo ] = table_clean
```

```python
all_tables
```

4. Append all the tables in just one DataFrame and export an excel file and save it in the folder [\_data\_results](https://github.com/alexanderquispe/Diplomado_PUCP/tree/main/Lecture_7/_data_results). 

```python
final_data = pd.concat( all_tables.values() ).reset_index( drop = True )
```

```python
final_data_ = final_data.rename(columns={'TOTAL DE VOTOS.2': 'political_parties', 'TOTAL DE VOTOS.3': 'total', 'TOTAL DE VOTOS.4': 'share_valid', 'TOTAL DE VOTOS.5':'share_emitied' })
final_data_
```

```python
def beta_OLS(x: pd.DataFrame, y: pd.Series, intercept = True):
    
    if not isinstance( x , pd.DataFrame ):
        raise TypeError( "X variable is not pd.DataFrame.")
        
    if not isinstance( y, pd.Series ):
        raise TypeError( "Y variable is not pd.Series.")
    
    if not isinstance( intercept, bool ):
        raise TypeError( " Intercept is not a bool")
    
    X = x.copy()
    if (intercept == True):
        intercept = np.ones( len(x))  
        X.insert(0, "Intercept", intercept, True)
    
    names = list(X.columns)

    result = np.linalg.inv(X.T @ X) @ (X.T @ y)
    result_1 = pd.DataFrame(result, index = names, columns = ['Coef.'])
    return result_1
```

```python
def var_OLS(x: pd.DataFrame, y: pd.Series, intercept  = True):
    
    if not isinstance( x , pd.DataFrame ):
        raise TypeError( "X variable is not pd.DataFrame.")
        
    if not isinstance( y, pd.Series ):
        raise TypeError( "Y variable is not pd.Series.")
    
    if not isinstance( intercept, bool ):
        raise TypeError( "Intercept is not a bool")
        
    X = x.copy()
    
    if (intercept == True):
        intercept = np.ones( len(x))
        X.insert(0, "Intercept", intercept, True)
    
    names = list(X.columns)

    result = np.linalg.inv(X.T @ X) @ (X.T @ y)

    # Defining errors and variance
    res = y - X @ result
    sigma_sqr = ( res.T @ res) / (len(y)-(len(result)))

    # getting var_beta_ols
    var_ols = sigma_sqr * np.linalg.inv(X.T @ X)
    
    var_ols = pd.DataFrame(var_ols, index = names, columns = names)
    return var_ols
```

```python
import scipy.stats as stats
```

```python
def reg_OLS(x: pd.DataFrame, y: pd.Series, intercept = True):
    
    if not isinstance( x , pd.DataFrame ):
        raise TypeError( "X variable is not pd.DataFrame type.")
        
    if not isinstance( y, pd.Series ):
        raise TypeError( "Y variable is not pd.Series type.")
    
    if not isinstance( intercept, bool ):
        raise TypeError( " Intercept is not a bool")
        
    X = x.copy()
    
    if (intercept == True):
        intercept = np.ones( len(x))
        X.insert(0, "Intercept", intercept, True)
   
        
    # Defining beta ols
    result = np.linalg.inv(X.T @ X) @ (X.T @ y)
    
    # Defining errors and variance
    res = y - X @ result
    sigma_sqr = ( res.T @ res) / (len(y)-(len(result)))

    # getting var_beta_ols
    var_ols = sigma_sqr * np.linalg.inv(X.T @ X) 
    
    # Standard errors of the estimated coefficients
    stderr = np.sqrt(np.diagonal(var_ols))

    # t values
    t_values = result/ stderr
    
    # P-values
    p_value = 2 * stats.t.cdf(x = -np.abs(t_values), df = len(y) - len(result))
    
    #  Calculate the confidence intervals:
    lower = result - stats.t.ppf(q = 1 - 0.05 / 2, df = len(y)-len(result)) * stderr
    upper = result + stats.t.ppf(q = 1 - 0.05 / 2, df = len(y)-len(result)) * stderr
    
    final_output = pd.DataFrame()
    final_output["Coef"],final_output["Std.Err."], final_output["t"], final_output["P>|t|"], final_output["[0.025"], final_output["0.975]"] = [result, stderr, t_values, p_value, lower, upper]
    
    #betas = ['Intercept','np.log(Q)', 'np.log(PL)', 'np.log(PF)', 'np.log(PK)']
    names = list(X.columns)
    final_output.index = names
    
    return final_output
```

```python
def restricted_OLS(y, X, L, r, intercept = True):
    
    # Import basic libraries
    import numpy as np
    import pandas as pd
    
    
    # Data validation
    if not isinstance( y , pd.Series ):
        raise TypeError( "y is not a pd.Series.")
        
    if not isinstance( X, pd.DataFrame ):
        raise TypeError( "X is not a pd.DataFrame.")
        
    if not isinstance( L, np.ndarray ):
        raise TypeError( "L is not a np.array.")  

    if not isinstance( r, np.ndarray ):
        raise TypeError( "r is not a np.array.")
    
    if not isinstance( intercept, bool ):
        raise TypeError( "Please select a valid type of intercept value: True or False")
        
    
    # Data as np to regression
    X_np_sub = np.asarray(X)
    y_np     = np.asarray(y)
    
    
    # Intercept matrix construction
    if intercept == True:
        # X matrix with Intercept
        intercept_np = np.ones(len(y))
        X_np         = np.c_[intercept_np, X_np_sub]
        # Variable Names with Intercept
        var_names    = list(X.columns)
        var_names.insert(0, "Intercept")
        # Restriction matrix with Intercept
        L_zeros      = np.zeros(len(r))
        L_np         = np.c_[L_zeros, L]
        
    elif intercept == False:
        # X matrix without Intercept
        X_np       = X_np_sub
        # Variable Names without Intercept
        var_names  = list(X.columns)
        # Restriction matrix without Intercept
        L_np       = L
        
    else:
        print("Please select a valid type of intercept value: True or False")
    
    # No scientific notation in dataframes
    pd.options.display.float_format = '{:.6f}'.format
    
    # Performing and storing regression
    Î²_ols = np.linalg.inv(X_np.T @ X_np) @ (X_np.T @ y_np)
    Î²_rls = Î²_ols - np.linalg.inv(X_np.T @ X_np) @ L_np.T @ np.linalg.inv( L_np @ np.linalg.inv(X_np.T @ X_np) @ L_np.T) @ (L_np @ Î²_ols - r )
    Î²_rls_pd = pd.DataFrame({'Coeficients': Î²_rls})
    Î²_rls_pd.index = var_names
    
    # Defining errors and variance
    res = y_np - X_np @ Î²_ols
    sigma_sqr = ( res.T @ res) / (len(y_np)-(len(X_np[0])+1))

    # getting var_beta_rls
    var_Î²_rls = sigma_sqr * np.linalg.inv(X_np.T @ X_np) - sigma_sqr * np.linalg.inv(X_np.T @ X_np) @ L_np.T @ np.linalg.inv( L_np @ np.linalg.inv(X_np.T @ X_np) @ L_np.T) @ L_np @ np.linalg.inv(X_np.T @ X_np)

    # showing var-cov coef. matrix
    vbr = pd.DataFrame(var_Î²_rls)
    vbr.columns = var_names
    vbr.index = var_names
    
    return Î²_rls_pd, vbr
```

```python
class RegClass( object ): #Building a class
    
    __slots__ = [ 'X', 'y', 'intercept', 'beta_OLS', 'var_OLS', 'rest_beta_OLS', 'rest_var_OLS', ] #Defining initial attributes 
    
    def __init__( self, X : pd.DataFrame, y : pd.Series, intercept = True ): #Defining outputs
        
        if not isinstance( X, pd.DataFrame ):
            raise TypeError( 'X is not a pd.DataFrame' )
    
        if not isinstance( y, pd.Series ):
            raise TypeError( 'y is not a pd.Series' )
    
        if not isinstance( intercept, bool ):
            raise TypeError( 'intercept is not a bool' )
        
        self.X = X
        self.y = y
        self.intercept = intercept
```

```python
def reg_OLS( self ): #Defining method reg_OLS
        
        self.beta_OLS = beta_OLS( self.X, self.y, intercept = self.intercept)
        self.var_OLS = var_OLS( self.X, self.y, intercept = self.intercept)
```

```python
def restricted_OLS( self, L : np.ndarray, r : np.ndarray, ): #Defining method restricted_OLS
        
        self.rest_beta_OLS = beta_RLS( self.X, self.y, L, r, intercept = self.intercept )
        self.rest_var_OLS = var_RLS( self.X, self.y, L, r, intercept = self.intercept )
```

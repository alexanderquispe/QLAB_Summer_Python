```python
def var_OLS ( X , y , intercept = True ): 
    
    if not isinstance( X ,  pd.DataFrame ):
            raise TypeError( "X  variable do not meet the requirements")

    if not isinstance( y ,  pd.Series ):
            raise TypeError( "y  variable do not meet the requirements")
            
    if (intercept == True) :
        X[ 'Intercept' ] = 1

        columns = X.columns.tolist()

        new_cols = [ columns[-1 ] ] + columns[0:-1]

        X = X.loc[ : , new_cols ]

    else: 
        pass
    
    index_names = X.columns.tolist()
    X_np = X.values
    y_np = y.values.reshape( -1, 1 )
        
    
    Œ≤_ols = np.linalg.inv(X_np.T @ X_np) @ (X_np.T @ y_np)
    res = y_np - X_np @ Œ≤_ols
    sigma_sqr = ( res.T @ res) / (len(y_np)-(len(X_np[0])+1))
    
    result_form = sigma_sqr * np.linalg.inv(X_np.T @ X_np) 
    result = pd.DataFrame(result_form)
    result.columns = index_names
    result.index = index_names
    result
    #result = pd.DataFrame( result_form , index = index_names , columns = ['Coef.'])
  
  
    return result
```

```python
def reg_OLS ( X, y, intercept = True ):
    
    if not isinstance( X,  pd.DataFrame ):
            raise TypeError( "X  variable do not meet the requirements." )

    if not isinstance( y,  pd.Series ):
            raise TypeError( "y variable do not meet the requirements." )
            
    if (intercept == True) :
        X[ 'Intercept' ] = 1
        
        columns = X.columns.tolist()
        new_cols = [ columns[-1 ] ] + columns[0:-1]
        X = X.loc[ : , new_cols ]
    
    else: 
        pass

    index_names = X.columns.tolist()

    coef = beta_OLS( X, y, intercept = True )       
    coef = coef.values
    
    var = var_OLS( X, y, intercept = True )       
    var = var.values
    
    st_error = np.sqrt(np.diag (var))
        
    
    t_value = coef.ravel()/np.sqrt( np.diag( var ))
    
    N_OBS = X.shape[ 0 ] 
    k = len(coef)
    
    p_tvalue = 2 * stats.t.cdf( x = - np.abs( t_value ), df =  N_OBS - k) 
    
    bound = st_error * stats.t.ppf( ( 1 + 0.95 ) / 2., N_OBS - k )
    up_bd = coef.ravel() + bound 
    lw_bd = coef.ravel() - bound

    result1 = { "Coef."    : coef.ravel(),  
                "Std.Err." : st_error.ravel(),
                "t"        : t_value.ravel(),
                "P>|t|"    : p_tvalue.ravel(),
                "[0.025"   : lw_bd.ravel(),
                "0.975]"   : up_bd.ravel()
                }
    
    result = pd.DataFrame( result1 , index = index_names )
    
    return result
```

```python
def restricted_OLS ( X, y, L, r, intercept = True ):
    
    if not isinstance( X ,  pd.DataFrame ):
            raise TypeError( "X variable do not meet the requirements." )

    if not isinstance( y ,  pd.Series ):
            raise TypeError( "y variable do not meet the requirements." )
            
    if not isinstance( L ,  np.ndarray ):
            raise TypeError( "L variable do not meet the requirements.")
            
    if not isinstance( r ,  np.ndarray ):
            raise TypeError( "r variable do not meet the requirements.")
            
    if (intercept == True) :
        X[ 'Intercept' ] = 1
        
        columns = X.columns.tolist()

        new_cols = [ columns[-1 ] ] + columns[0:-1]

        X = X.loc[ : , new_cols ]

    else:
        pass
    
    index_names = X.columns.tolist()
    X_np = X.values
    y_np = y.values.reshape( -1, 1 )
   
    

    # beta OLS
    
    coef = beta_OLS( X, y, intercept = True ).values
    #coef  = np.linalg.inv(X_np.T @ X_np) @ (X_np.T @ y_np)
    
    # beta RLS
    Œ≤_rls = coef - np.linalg.inv(X_np.T @ X_np) @ L.T @ np.linalg.inv( L @ np.linalg.inv(X_np.T @ X_np) @ L.T) @ (L @ coef - r )

    # Var
    var = var_OLS ( X, y, intercept = True ).values
    var_Œ≤_rls = var - var @ L.T @ np.linalg.inv( L @ np.linalg.inv(X_np.T @ X_np) @ L.T) @ L @ np.linalg.inv(X_np.T @ X_np)
   # var_Œ≤_rls = var_Œ≤_rls.values
    
    desv = np.sqrt( np.diag( var_Œ≤_rls))
   # desv = desv.values

    t_value = coef.ravel()/np.sqrt( np.diag( var_Œ≤_rls )) 

    N_OBS = X.shape[ 0 ]
    k = len(coef)
    P = r.shape [ 0 ] 
    p_tvalue = 2 * stats.t.cdf( x = - np.abs( t_value ), df =  N_OBS - k - P) 
    
    bound = st_error * stats.t.ppf( ( 1 + 0.95 ) / 2., N_OBS - k - P)
    up_bd = coef.ravel() + bound 
    lw_bd = coef.ravel() - bound
   

    result1 = { "Coef."    : Œ≤_rls.ravel(),  
                "Std.Err." : desv.ravel(),
                "t"        : t_value.ravel(),
                "P>|t|"    : p_tvalue.ravel(),
                "[0.025"   : lw_bd.ravel(),
               "0.975]"   : up_bd.ravel()
                }

    result = pd.DataFrame( result1 , index = columns )

    return result
```

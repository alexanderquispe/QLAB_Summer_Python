```python
# Defining beta_OLS function  

def beta_ols( X:pd.DataFrame , y:pd.Series, intercept:bool = True):
    
    if not isinstance( X , pd.DataFrame ):
        raise TypeError( "X is not a pd.DataFrame type." )
        
    if not isinstance( y , pd.Series ):
        raise TypeError( "y is not a pd.Series type." )
        
    if not isinstance( intercept , bool ):
        raise TypeError( "intercept is not a boolean type." )
    
    if intercept == True:
        X_new = X.copy() 
        X_new.insert(0, 'intercept', 1)
        beta_ols = np.linalg.inv(X_new.T @ X_new) @ X_new.T @ y 
        beta_ols = pd.DataFrame(beta_ols)
        beta_ols[' '] = X_new.columns.tolist()
        beta_ols.set_index(' ', inplace=True)
        beta_ols.rename(columns = {0:'Coef.'}, inplace=True)
        results = beta_ols
    
    else:
        X_new = X.copy() 
        beta_ols = np.linalg.inv(X_new.T @ X_new) @ X_new.T @ y 
        beta_ols = pd.DataFrame(beta_ols)
        beta_ols[' '] = X_new.columns.tolist()
        beta_ols.set_index(' ', inplace=True)
        beta_ols.rename(columns = {0:'Coef.'}, inplace=True)
        results = beta_ols 
    
    return results
```

```python
# Defining var_OLS function  

def var_ols( X:pd.DataFrame , y:pd.Series, intercept:bool = True) -> pd.DataFrame:
    
    if not isinstance( X , pd.DataFrame ):
        raise TypeError( "X is not a pd.DataFrame type." )
        
    if not isinstance( y , pd.Series ):
        raise TypeError( "y is not a pd.Series type." )
        
    if not isinstance( intercept , bool ):
        raise TypeError( "intercept is not a boolean type." )
    
    if intercept == True:
        X_new = X.copy() 
        X_new.insert(0, 'intercept', 1) 
        names = X_new.columns.tolist()
        beta_est = beta_ols( X, y, intercept)   ### Estimating parameters
        y_pred = X_new @ beta_est               ### Calculating predicted values
        y_pred = y_pred.squeeze()
        n_data = X_new.shape[0]
        k_param = X_new.shape[1]
        sr2 = ( (1 / (n_data - k_param)) * (y_pred - y).T  @ (y_pred - y)) ### Calculating the variance of the regression
        var_beta = sr2*np.linalg.inv(X_new.T @ X_new)                ### Getting variance and covariance Matrix
        var_beta = pd.DataFrame(var_beta, columns=names)
        var_beta[' '] = names
        var_beta.set_index(' ', inplace=True)
        
    else:
        X_new = X.copy() 
        names = X_new.columns.tolist()
        beta_est = beta_ols( X, y, intercept)   ## Estimating parameters
        y_pred = X_new @ beta_est               ## Calculating predicted values
        y_pred = y_pred.squeeze()
        n_data = X_new.shape[0]
        k_param = X_new.shape[1]
        sr2 = ( (1 / (n_data - k_param)) * (y_pred - y).T  @ (y_pred - y)) ## Calculating the variance of the regression
        var_beta = sr2*np.linalg.inv(X_new.T @ X_new)  #### Getting variance and covariance Matrix
        var_beta = pd.DataFrame(var_beta, columns=names)
        var_beta[' '] = names
        var_beta.set_index(' ', inplace=True)
        
    results = var_beta
   
    return results
```

```python
# Defining reg_OLS function  

from scipy.stats import t
def reg_ols( X:pd.DataFrame , y:pd.Series, intercept:bool = True) -> pd.DataFrame:

    if not isinstance( X , pd.DataFrame ):
        raise TypeError( "X is not a pd.DataFrame type." )
        
    if not isinstance( y , pd.Series ):
        raise TypeError( "y is not a pd.Series type." )
        
    if not isinstance( intercept , bool ):
        raise TypeError( "intercept is not a boolean type." )
    
    if intercept == True:
        X_new = X.copy() 
        X_new.insert(0, 'intercept', 1) 
        names = X_new.columns.tolist()
        n_data = X_new.shape[0]
        k_param = X_new.shape[1]
        
        beta_est = beta_ols( X, y, intercept )   ### Estimating parameters
        beta_est.reset_index(inplace=True)
        var_beta = var_ols( X, y, intercept )   ### Estimating variance and covariance matrix
        stand_error = pd.DataFrame(np.sqrt(np.diag(var_beta)), columns= ['Std.Err'])  ## Extracting standard error of parameters
        resume_df = pd.concat([beta_est, stand_error], axis=1)
        resume_df['t'] = resume_df['Coef.']/resume_df['Std.Err']     ### Computing t value
        resume_df['P>|t|'] = 1 - 2*np.abs(0.5 - np.vectorize(t.cdf)(resume_df['t'], n_data-k_param-1))    ### Computing p-value
        resume_df['0.025'] = resume_df['Coef.'] - resume_df['Std.Err']*t.ppf( q = 1 - 0.05 / 2, df = n_data - k_param ) ### Computing lower interval
        resume_df['0.975'] = resume_df['Coef.'] + resume_df['Std.Err']*t.ppf( q = 1 - 0.05 / 2, df = n_data - k_param ) ### Computing upper interval
        resume_df[' '] = names
        resume_df.set_index(' ', inplace=True)
        
    else:
        X_new = X.copy() 
        names = X_new.columns.tolist()
        n_data = X_new.shape[0]
        k_param = X_new.shape[1]
        
        beta_est = beta_ols( X, y, intercept )   ### Estimating parameters
        beta_est.reset_index(inplace=True)
        var_beta = var_ols( X, y, intercept )   ### Estimating variance and covariance matrix
        stand_error = pd.DataFrame(np.sqrt(np.diag(var_beta)), columns= ['Std.Err'])  ## Extracting standard error of parameters
        resume_df = pd.concat([beta_est, stand_error], axis=1)
        resume_df['t'] = resume_df['Coef.']/resume_df['Std.Err']     ### Computing t value
        resume_df['P>|t|'] = 1 - 2*np.abs(0.5 - np.vectorize(t.cdf)(resume_df['t'], n_data-k_param-1))    ### Computing p-value
        resume_df['0.025'] = resume_df['Coef.'] - resume_df['Std.Err']*t.ppf( q = 1 - 0.05 / 2, df = n_data - k_param ) ### Computing lower interval
        resume_df['0.975'] = resume_df['Coef.'] + resume_df['Std.Err']*t.ppf( q = 1 - 0.05 / 2, df = n_data - k_param ) ### Computing upper interval
        resume_df[' '] = names
        resume_df.set_index(' ', inplace=True)
        
    results = resume_df
   
    return results
```

```python
# Defining restricted_OLS function  

from scipy.stats import t
def restricted_OLS( X:pd.DataFrame , y:pd.Series, L:np.array, r:np.array, intercept:bool = True) -> pd.DataFrame:

    if not isinstance( X , pd.DataFrame ):
        raise TypeError( "X is not a pd.DataFrame type." )
        
    if not isinstance( y , pd.Series ):
        raise TypeError( "y is not a pd.Series type." )
        
    if not isinstance( L ,np.ndarray ):
        raise TypeError( "L is not a array type." )
        
    if not isinstance( r , np.ndarray):
        raise TypeError( "r is not a array type." )
        
    if not isinstance( intercept , bool ):
        raise TypeError( "intercept is not a boolean type." )
    
    if intercept == True:
        X_new = X.copy() 
        X_new.insert(0, 'intercept', 1) 
        L_new = L.copy() 
        L_new = np.append([[0],[0],[0],[0],[0]],L_new, axis = 1)
        r_new = r.copy() 
        names = X_new.columns.tolist()
        n_data = X_new.shape[0]
        k_param = X_new.shape[1]
        M_restric = L_new.shape[0]
        
        # RLS estimation
        beta_est = beta_ols( X, y, intercept )   ### Estimating parameters using OLS
        beta_est = beta_est.squeeze()
        beta_est_restric = beta_est - np.linalg.inv(X_new.T @ X_new) @ L_new.T@ np.linalg.inv(L_new @ (np.linalg.inv(X_new.T @ X_new)) @ L_new.T)@(L_new @ beta_est - r_new )   
        beta_est_restric = pd.DataFrame(beta_est_restric)
        
        # Variance and covariance Matrix of RLS estimation
        y_pred = X_new @ beta_est_restric               ### Calculating predicted values
        y_pred = y_pred.squeeze()
        sr2 = ( (1 / (n_data - (k_param + 1 - M_restric))) * (y_pred - y).T  @ (y_pred - y)) ### Calculating the variance of the regression
        var_rls_beta = sr2*np.linalg.inv(X_new.T @ X_new) - sr2*np.linalg.inv(X_new.T @ X_new)@ L_new.T@ np.linalg.inv((L_new @ np.linalg.inv(X_new.T @ X_new) @ L_new.T)) @ L_new @ np.linalg.inv(X_new.T @ X_new)  ### Getting variance and covariance Matrix
        var_rls_beta = pd.DataFrame(var_rls_beta, columns=names)
        var_rls_beta[' '] = names
        var_rls_beta.set_index(' ', inplace=True)
        
        beta_est_restric.reset_index(inplace=True)
        stand_error = pd.DataFrame(np.sqrt(np.diag(var_rls_beta)), columns= ['Std.Err'])  ## Extracting standard error of parameters
        resume_df = pd.concat([beta_est_restric, stand_error], axis=1)
        resume_df['t'] = resume_df['Coef.']/resume_df['Std.Err']     ### Computing t value
        resume_df['P>|t|'] = 1 - 2*np.abs(0.5 - np.vectorize(t.cdf)(resume_df['t'], n_data-k_param-1))    ### Computing p-value
        resume_df['0.025'] = resume_df['Coef.'] - resume_df['Std.Err']*t.ppf( q = 1 - 0.05 / 2, df = n_data - k_param ) ### Computing lower interval
        resume_df['0.975'] = resume_df['Coef.'] + resume_df['Std.Err']*t.ppf( q = 1 - 0.05 / 2, df = n_data - k_param ) ### Computing upper interval
        resume_df[' '] = names
        resume_df.set_index(' ', inplace=True)
        
    else:
        X_new = X.copy() 
        L_new = L.copy() 
        r_new = r.copy() 
        names = X_new.columns.tolist()
        n_data = X_new.shape[0]
        k_param = X_new.shape[1]
        M_restric = L_new.shape[0]
        
        # RLS estimation
        beta_est = beta_ols( X, y, intercept )   ### Estimating parameters using OLS
        beta_est = beta_est.squeeze()
        beta_est_restric = beta_est - np.linalg.inv(X_new.T @ X_new) @ L_new.T@ np.linalg.inv(L_new @ (np.linalg.inv(X_new.T @ X_new)) @ L_new.T)@(L_new @ beta_est - r_new )   
        beta_est_restric = pd.DataFrame(beta_est_restric)
        
        # Variance and covariance Matrix of RLS estimation
        y_pred = X_new @ beta_est_restric               ### Calculating predicted values
        y_pred = y_pred.squeeze()
        sr2 = ( (1 / (n_data - (k_param + 1 - M_restric))) * (y_pred - y).T  @ (y_pred - y)) ### Calculating the variance of the regression
        var_rls_beta = sr2*np.linalg.inv(X_new.T @ X_new) - sr2*np.linalg.inv(X_new.T @ X_new)@ L_new.T@ np.linalg.inv((L_new @ np.linalg.inv(X_new.T @ X_new) @ L_new.T)) @ L_new @ np.linalg.inv(X_new.T @ X_new)  ### Getting variance and covariance Matrix
        var_rls_beta = pd.DataFrame(var_rls_beta, columns=names)
        var_rls_beta[' '] = names
        var_rls_beta.set_index(' ', inplace=True)
        
        beta_est_restric.reset_index(inplace=True)
        stand_error = pd.DataFrame(np.sqrt(np.diag(var_rls_beta)), columns= ['Std.Err'])  ## Extracting standard error of parameters
        resume_df = pd.concat([beta_est_restric, stand_error], axis=1)
        resume_df['t'] = resume_df['Coef.']/resume_df['Std.Err']     ### Computing t value
        resume_df['P>|t|'] = 1 - 2*np.abs(0.5 - np.vectorize(t.cdf)(resume_df['t'], n_data-k_param-1))    ### Computing p-value
        resume_df['0.025'] = resume_df['Coef.'] - resume_df['Std.Err']*t.ppf( q = 1 - 0.05 / 2, df = n_data - k_param ) ### Computing lower interval
        resume_df['0.975'] = resume_df['Coef.'] + resume_df['Std.Err']*t.ppf( q = 1 - 0.05 / 2, df = n_data - k_param ) ### Computing upper interval
        resume_df[' '] = names
        resume_df.set_index(' ', inplace=True)
        
    results = resume_df
   
    return results, beta_est_restric, var_rls_beta
```

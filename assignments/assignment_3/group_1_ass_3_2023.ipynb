```python
##EXERCISE 2
#We create a np.arange from 0 to 8 (as the last number will not get accounted, we put 9)
threexthree = np.arange(0, 9) 
#Now, we divide the np.arange using the function reshape to get the form of a 3x3.
threexthree=threexthree.reshape(3, 3) 
print(threexthree)
```

```python
##EXERCISE 3
#The code used works as follow:
#We wish to get a group of four values. So, the first parenthesis goes from 1 to 5. We remember that 5 will then be excluded, because it is the maximum and values just go in a step of 1 by 1).
#After, we get a [None, :]. This, united to the [:, None] at last. Both mark the beginning and end of the array. The ":" show the continuity of the array, as we just need to create an array 1-dimensional.
#Then, it appears a +1. This means that every value of every row of this np.arange will be added by 1, realted to their previous row.
#Also, it is specified the number "7", because to arrive at the [ 7,  8,  9, 10] asked, we need the first arrangement of 4 values repeated 7 times. Another way to interpretate this is that the first value of the first row (1) need to be multiplied by 7 times so it arrives to the number "7" at the beginning of the group.

Z = np.arange(1,5)[None, :] + 1*np.arange(7)[:, None]
Z
```

```python
#We use the previous code to concatenate the value using "np.concatenate" and we create the object R.

R=np.concatenate(np.arange(1,5)[None, :] + 1*np.arange(7)[:, None])
R
```

```python
##EXERCISE 5
np.nan == np.nan
#This expression results false because Nan has two meaning at Python. 
#At the beginning of this file, we specified that np was numpy. 
#This means that NAN, in one hand, is a float value, which is a value that is not defined yet, but will get (sooner or later) a number, text, etc. 
#Nevertheless, on the other hand, Nan is "Not a number". Here, it says that the value is unexistant.
#So, in case we affirm that these two values are equal, that would be false. As one value is not defined yet, we can't say it is equal to a value that is already a blanck space. 
#In other words, a value available to fill is not the same to one that is already defined to be empty.
```

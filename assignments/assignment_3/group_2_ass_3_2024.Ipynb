```python
# Extraer el aÃ±o usando regex
df_long['Year_operacion'] = df_long['LCU_year'].str.extract(r'(\d{4})')

# Convertir la columna 'Year' a tipo numÃ©rico
df_long['Year_operacion'] = pd.to_numeric(df_long['Year_operacion'], errors='coerce')
```

```python
df_long.head()
```

```python
df_long['Operating_revenue'] = df_long['Operating_revenue'].replace('n.a.', pd.NA)
df_long = df_long.dropna(subset=['Operating_revenue'])
```

```python
df_long.head()
```

```python
def calling_data(firm_name, start_year, end_year=None):
    # Filtrar el DataFrame por nombre de empresa y rango de aÃ±os
    filtered_df = df_long[df_long['Firm Name'].isin(firm_name)]
    
    if end_year:
        # Si se proporciona un aÃ±o de fin, filtrar por rango de aÃ±os
        filtered_df = filtered_df[(filtered_df['Last avail. year'] >= start_year) & (filtered_df['Last avail. year'] <= end_year)]
    else:
        # Si no se proporciona un aÃ±o de fin, filtrar por aÃ±o de inicio
        filtered_df = filtered_df[filtered_df['Year_operacion'] == start_year]

    # Devolver el DataFrame resultante
    return filtered_df, column_definitions

```

```python
# Llamar a la funciÃ³n con tu DataFrame y obtener el resultado
resulting_df, column_definitions = calling_data(df_long['Firm Name'], df_long['Year_operacion'])
```

```python
class DataProcessor:
    def __init__(self, df_long, column_definitions):
        self.df = df_long
        self.column_definitions = column_definitions

    def calling_data(self, firm_name, start_year, end_year=None):
        filtered_df = self.df[self.df['Firm Name'].isin(firm_name)]
        
        if end_year:
            filtered_df = filtered_df[(filtered_df['Last avail. year'] >= start_year) & (filtered_df['Last avail. year'] <= end_year)]
        else:
            filtered_df = filtered_df[filtered_df['Year_operacion'] == start_year]

        return filtered_df, self.column_definitions

    def descriptive(self, column_name):
        if column_name not in self.df.columns:
            raise ValueError(f"Column '{column_name}' not found in DataFrame.")
        
        column_data = self.df[column_name]

        # Calcular estadÃ­sticas descriptivas
        mean_value = column_data.mean()
        max_value = column_data.max()
        min_value = column_data.min()
        median_value = column_data.median()

        # Calcular el promedio mÃ³vil (rolling average) con ventana de 3
        rolling_average = column_data.rolling(window=3).mean()

        return {
            'mean': mean_value,
            'max': max_value,
            'min': min_value,
            'median': median_value,
            'rolling_average': rolling_average
        }
```

```python
data_processor = DataProcessor(df_long, column_definitions)
```

```python
resulting_df, resulting_column_definitions = data_processor.calling_data(df_long['Firm Name'], df_long['Year_operacion'])
```

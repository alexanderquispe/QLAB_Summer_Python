.....Procesamiento adicional de las columnas 'Año' y 'Tipo de Ingreso':
Divide la columna 'Año' para obtener solo el año y actualiza la columna 'Año' en consecuencia.
Agrega una nueva columna llamada 'Tipo de Ingreso', que se deriva de la columna 'Año' al eliminar la última palabra.


.....Adición de una columna 'diccionario':
Agrega una nueva columna llamada 'diccionario', cuyos valores se obtienen aplicando la función get_definition al año y al tipo de ingreso correspondientes.


```python
def calling_data(df, firm_name, start_year, end_year=None):

    if isinstance(firm_name, list):
        df = df[df['Company name Latin alphabet'].isin(firm_name)]
    else:
        df = df[df['Company name Latin alphabet'] == firm_name]
    
    if end_year:
        selected_columns = [col for col in df.columns if start_year <= get_year_from_column(col) <= end_year]
    else:
        selected_columns = [col for col in df.columns if start_year <= get_year_from_column(col)]
    new_df = df[['Company name Latin alphabet'] + selected_columns]
    
    new_df = new_df.melt(id_vars=['Company name Latin alphabet'], 
                         value_vars=selected_columns,
                         var_name='Año', value_name='Total')
    new_df['Tipo de Ingreso'] = new_df['Año'].apply(lambda x: x.split()[:-1])
    new_df['Año'] = new_df.apply(lambda row: get_year_from_column(row['Año']), axis=1)
    new_df['diccionario'] = new_df.apply(lambda row: get_definition(row['Año'], row['Tipo de Ingreso']), axis=1)
    return new_df
```

Esta función, get_year_from_column, devuelve el año de una columna dada:

Si la columna es de tipo entero, se devuelve tal cual.
Si no es un entero, intenta dividir la columna en palabras y devuelve el último elemento como un entero (suponiendo que sea una representación del año).
Si hay un error al intentar convertir la última palabra en un entero, devuelve -1.

```python
def get_year_from_column(column):
    if isinstance(column, int):
        return column
    try:
        return int(column.split()[-1])
    except ValueError:
        return -1
```

Esta función, get_definition, genera una definición para un año y un tipo de ingreso dados:

Obtiene la última palabra de la lista tipo_ingreso (o una cadena vacía si tipo_ingreso está vacío).
Devuelve una cadena que combina la última palabra del tipo de ingreso con el año en una frase descriptiva.
En resumen, la función construye una descripción de ingresos operativos con la última palabra del tipo de ingreso y el año proporcionados.

```python
def get_definition(year, tipo_ingreso):
    last_word = tipo_ingreso[-1] if tipo_ingreso else ''
    return f"Ingresos operativos moneda {last_word} con año {year}"
```

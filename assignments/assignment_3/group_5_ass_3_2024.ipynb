```python
import pandas as pd

def calling_data(firm_name, start_year, end_year=None):
    # Cargar el conjunto de datos desde la segunda hoja
    sheet_name = 'Results'
    file_path = 'C:/Users/a_aba/OneDrive/Escritorio/bogota_firms.xlsx'
    df = pd.read_excel(file_path, sheet_name=sheet_name) 

    # Filtrar columnas que contienen "Dollars"
    df = df.loc[:, ~df.columns.str.contains('USD')]

    # Seleccionar columnas desde 'Operating revenue (Turnover) th LCU 2023' hasta el final 
    selected_columns = df.columns[df.columns.get_loc('Operating revenue (Turnover)\nth LCU 2023'):]
    df_long = pd.melt(df, id_vars=['Company name Latin alphabet'], value_vars=selected_columns, var_name='Operating Revenue', value_name='Value')

    # Renombrar las columnas en el DataFrame
    rename_dict = {}
    for year in range(2011, 2024):
        rename_dict[f'Operating revenue (Turnover)\nth LCU {year}'] = f'Rev{year}'

    rename_dict.update({
        'Company name Latin alphabet': 'FirmName',
        'Country ISO code': 'ISO Code',
        'City\nLatin Alphabet': 'City Name',
        'NACE Rev. 2, core code (4 digits)': 'NACE Rev. 2, core',
        'Consolidation code': 'Consolid. Code',
        'Last avail. year': 'Last year'
    })

    df_long.rename(columns=rename_dict, inplace=True)

    # Crear un diccionario de definiciones de columnas
    columns_definition = {
        'FirmName': 'Name of the firm',
        'ISO Code': 'Country ISO code',
        'City Name':'City\nLatin Alphabet',
        'NACE Rev. 2, core': 'NACE Rev. 2, core code (4 digits)',
        'Consolid. Code': 'Consolidation code',
        'Last year': 'Last avail. year',
    }

    # Filtrar el DataFrame segÃºn el nombre de la empresa
    if isinstance(firm_name, list):
        filtered_df = df_long[df_long['FirmName'].isin(firm_name)]
    else:
        filtered_df = df_long[df_long['FirmName'] == firm_name]

    # Filtrar el DataFrame segÃºn el rango de aÃ±os
    if end_year is not None:
        filtered_df = filtered_df[(filtered_df['Operating Revenue'] >= f'Rev{start_year}') & (filtered_df['Operating Revenue'] <= f'Rev{end_year}')]
    else:
        filtered_df = filtered_df[filtered_df['Operating Revenue'].str.contains(f'Rev{start_year}')]
    
    print(filtered_df)

    return filtered_df, columns_definition


# Ejemplo de uso
result_df, result_columns_definition = calling_data('BACKBONE COMUNICACIONES LTDA', 2020, 2022)
result_df.head()
print(result_columns_definition)
```

```python
import pandas as pd

class DataProcessor:
    def __init__(self, file_path):
        self.file_path = file_path
        self.columns_definition = {
            'FirmName': 'Name of the firm',
            'ISO Code': 'Country ISO code',
            'City Name':'City\nLatin Alphabet',
            'NACE Rev. 2, core': 'NACE Rev. 2, core code (4 digits)',
            'Consolid. Code': 'Consolidation code',
            'Last year': 'Last avail. year',
            'Rev2023': 'Operating revenue (Turnover) th LCU 2023',
            'Rev2022': 'Operating revenue (Turnover) th LCU 2022',
            'Rev2021': 'Operating revenue (Turnover) th LCU 2021',
            'Rev2020': 'Operating revenue (Turnover) th LCU 2020',
            'Rev2019': 'Operating revenue (Turnover) th LCU 2019',
            'Rev2018': 'Operating revenue (Turnover) th LCU 2018',
            'Rev2017': 'Operating revenue (Turnover) th LCU 2017',
            'Rev2016': 'Operating revenue (Turnover) th LCU 2016',
            'Rev2015': 'Operating revenue (Turnover) th LCU 2015',
            'Rev2014': 'Operating revenue (Turnover) th LCU 2014',
            'Rev2013': 'Operating revenue (Turnover) th LCU 2013',
            'Rev2012': 'Operating revenue (Turnover) th LCU 2012',
            'Rev2011': 'Operating revenue (Turnover) th LCU 2011',
        }
        self.df = self.load_and_process_data()

    def load_and_process_data(self):
        # Cargar el conjunto de datos desde la segunda hoja
        sheet_name = 'Results'
        df = pd.read_excel(self.file_path, sheet_name=sheet_name) 

        # Filtrar columnas que contienen "Dollars"
        df = df.loc[:, ~df.columns.str.contains('USD')]

        # Seleccionar columnas desde 'Operating revenue (Turnover) th LCU 2023' hasta el final 
        selected_columns = df.columns[df.columns.get_loc('Operating revenue (Turnover)\nth LCU 2023'):]
        df_long = pd.melt(df, id_vars=['Company name Latin alphabet'], value_vars=selected_columns, var_name='Operating Revenue', value_name='Value')

        # Renombrar las columnas en el DataFrame
        rename_dict = {}
        for year in range(2011, 2024):
            rename_dict[f'Operating revenue (Turnover)\nth LCU {year}'] = f'Rev{year}'

        rename_dict.update({
            'Company name Latin alphabet': 'FirmName',
            'Country ISO code': 'ISO Code',
            'City\nLatin Alphabet': 'City Name',
            'NACE Rev. 2, core code (4 digits)': 'NACE Rev. 2, core',
            'Consolidation code': 'Consolid. Code',
            'Last avail. year': 'Last year'
        })

        df_long.rename(columns=rename_dict, inplace=True)
        return df_long

    def descriptive(self):
        # Limpiar valores no numÃ©ricos
        self.df['Value'] = pd.to_numeric(self.df['Value'], errors='coerce')

        # Calcular estadÃ­sticas descriptivas
        descriptive_stats = self.df['Value'].describe()
        
        # Calcular la mediana
        median_value = self.df['Value'].median()

        # Calcular el promedio mÃ³vil
        rolling_average = self.df['Value'].rolling(window=3).mean()

        return {
            'Mean': descriptive_stats['mean'],
            'Max': descriptive_stats['max'],
            'Min': descriptive_stats['min'],
            'Median': median_value,
            'Rolling Average': rolling_average
        }


# Ejemplo de uso
data_processor = DataProcessor('C:/Users/a_aba/OneDrive/Escritorio/bogota_firms.xlsx')
result_descriptive = data_processor.descriptive()
print(result_descriptive)
```

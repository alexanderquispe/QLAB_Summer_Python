```python
# QUESTION 1.1 
# Inspect this dataset ./_data/bogota_firms.xlsx 
import pandas as pd
file_path = r"C:\Users\LENOVO\Desktop\Github\QLAB_Summer_Python\_data\bogota_firms.xlsx"

#select the corresponding sheet
df_Results = pd.read_excel(file_path, sheet_name="Results")

#View the kind of data
#print(df_Results.head())

selected_columns = df_Results.iloc[:, 7:33]
print(selected_columns)

# Convert the selected columns to numeric, coercing errors to NaN
selected_columns_numeric = selected_columns.apply(pd.to_numeric, errors='coerce')
```

```python
# 1.2 and show some descriptive statistics of it.
# Calculate the mean, max, and min along the specified axis (axis=1 for row-wise operations)
row_means = selected_columns_numeric.mean(axis=1)
row_max = selected_columns_numeric.max(axis=1)
row_min = selected_columns_numeric.min(axis=1)
row_std = selected_columns_numeric.std(axis=1)

# Create a new DataFrame with the results
result_df = pd.DataFrame({
    'Mean': row_means,
    'Max': row_max,
    'Min': row_min,
    'Std': row_std
})

# Now you have a DataFrame with the mean, max, and min values for each row
print(result_df)

```

```python
# QUESTION NUMBER 2. 
# Reshape the dataset from wide to long at year level. 
# Drop columns if they are in Dollars. Keep columns from Operating revenue (Turnover) th LCU 2023 to the end, 
# and the name of the firm.

# Specify the columns to drop based on the conditions
columns_to_drop = [col for col in df_Results.columns if 'USD' in col]

# Drop the specified columns
df_filtered = df_Results.drop(columns=columns_to_drop)

# Specify the columns to keep based on the conditions
columns_to_keep = ['Company name Latin alphabet'] + [col for col in df_filtered.columns if 'Operating revenue (Turnover)' in col]

# Select the relevant subset of columns
subset_df = df_filtered[columns_to_keep]

# Display the DataFrame after dropping and selecting columns
print(subset_df)
```

```python
# Use melt to reshape the DataFrame to long format
long_df = pd.melt(df_Results, id_vars='Company name Latin alphabet', value_vars=columns_to_keep[1:])

print(long_df)
```

```python
# QUESTION NUMBER 3. 
# Rename the columns and use short names for each column. This new name should be shorter than 20 characters.
# Get the original column names
original_columns = subset_df.columns

# Create a dictionary to map the old column names to the new ones
column_mapping = {original_columns[0]: 'Company_name'}
column_mapping.update({col: 'OR_TO_{}'.format(col[-4:]) for col in original_columns[1:]})

# Rename the columns using the mapping
df = subset_df.rename(columns=column_mapping)

# Display the DataFrame with the new column names
print(df)
```

```python
# QUESTION NUMBER 6
# Generate a class that do the same as the function, but has a method named as `descripitive` that 
# gives you the mean, max, min, median, and rolling average .

import pandas as pd

class DataFrameAnalyzer:  
    def __init__(self, dataframe):   #atributtes
        self.dataframe = dataframe

    def descriptive(self, window_size=None):
        # Select columns from index 8 to 32
        selected_columns = self.dataframe.iloc[:, 8:33]

        # Convert selected columns to numeric, coercing errors to NaN
        selected_columns_numeric = selected_columns.apply(pd.to_numeric, errors='coerce')

        # Calculate statistics along rows
        row_means = selected_columns_numeric.mean(axis=1)
        row_max = selected_columns_numeric.max(axis=1)
        row_min = selected_columns_numeric.min(axis=1)
        row_std = selected_columns_numeric.std(axis=1)

        # Create a new DataFrame with the results
        result_df = pd.DataFrame({
            'Mean': row_means,
            'Max': row_max,
            'Min': row_min,
            'Std': row_std
        })

        # Calculate mean of the years
        column_means = selected_columns_numeric.mean(axis=0)

        # Find the column with the best mean
        best_mean_column = column_means.idxmax()
        worst_mean_column = column_means.idxmin()

        # Display the results
        print("\nDescriptive Statistics (Row-wise):")
        print(result_df)

        print("\nMean of each column:")
        print(column_means)

        print("\nColumn with the best mean:", best_mean_column)
        print("Column with the worst mean:", worst_mean_column)

        if window_size is not None:
            # Calculate rolling average using a window of specified size
            rolling_avg = selected_columns_numeric.rolling(window=window_size, axis=1).mean()
            print("\nRolling Average:")
            print(rolling_avg)

file_path = r"C:\Users\Hp\Documents\GitHub\QLAB_Summer_Python\_data\bogota_firms.xlsx"
df_Results = pd.read_excel(file_path, sheet_name="Results")

# Instantiate the class and perform descriptive statistics
analyzer = DataFrameAnalyzer(df_Results)
analyzer.descriptive(window_size=3)
```

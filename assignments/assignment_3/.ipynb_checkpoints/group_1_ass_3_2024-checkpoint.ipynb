2. Reshape the dataset from wide to long at year level. Drop columns if they are in Dollars. Keep columns from Operating revenue (Turnover) th LCU 2023 to the end, and the name of the firm.

First we will remove the columns that have operating revenues in dollars.

```python
# We get the index of the first column of the set of columns that we want to extract.
a=list(df.columns).index('Operating revenue (Turnover)\nth LCU 2023')

# We use 'a' to create a list containing the names of the columns we are interested in.
b=list(df.columns[a:])

# We printed the list to see what names we got.
print(b)
```

```python
# You can see that the even-numbered elements of the list contain the names of the columns we want to delete. Then, we extract the even elements in the following list.
b1=b[1::2]

# With the 'drop' method we eliminate the columns whose name is in the previous list created.
df1=df.drop(columns=b1)

# And with the following we can see that the columns denominated in dollars have been eliminated:
list(df1.columns[a:])
```

Now with the following loop we rename each year's operating profits as the year they were recorded. We do this to abbreviate the names (which only differ by year).

```python
n,k=df1.shape # With this we make 'n' contain the total number of rows and 'k' the total number of columns.

for x in range(a,k):
    y=df1.columns[x][-4:] # With this we extract the last 4 strings of the column name, which conveniently forms the year of record.
    df1.rename(columns={df1.columns[x]:y},inplace=True) # With this we rename each column. Now they are named only by the year in which they were registered.
```

We now reshape the data set from width to length at the year level. For this we use `pd.melt`

```python
# We use the first column (which is conveniently an index) as the identifier of each observation. That's why 'id_vars=[df1.columns[0]]'
# value_vars contains the names of the columns to be reshaped.
# var_name is the column that will contain the names of the collapsed columns. We call it 'Year' because the only difference between the data of column names is the year (and it only contains the years).
# The observations of each column are reshaped into a single column called 'OR', which stands for 'operating revenue' for short.

df2=pd.melt(df1, id_vars=[df1.columns[0]], value_vars=list(df1.columns[a:k]), var_name='Year',value_name='OR') 

# We see the result.
df2
```

As you can see, a Dataframe long at year level has been generated. However, it is missing the column that indicates the name of the company to which each observation corresponds (and it is the only column that the problem also asks to maintain). So, we implement the following:

```python
df3=df[df.columns[0:2]] # We extract the first and second columns from the original Dataframe.

# Now we do the following merge. The first column serves as the identifier of each observation (that's why 'on=df1.columns[0]').
df4=pd.merge(df2, df3, on=df1.columns[0], how='left')
df4
```

Now we make the following arrangements to the DataFrame to improve its visualization:

```python
# We delete the first column since it is no longer useful to us.
df4.drop(columns=df4.columns[0],inplace=True)

# We reorder the columns as follows:
order=[df4.columns[2],df4.columns[0],df4.columns[1]]
df4=df4[order]

# We sort the data according to the year in ascending order.
df4=df4.sort_values(by='Year', ascending=True)

# We reset the index.
df4=df4.reset_index(drop=True)

# We convert the year data from string format to numpy format. This will be useful for the following questions as we will see.
df4['Year']=pd.DataFrame(np.float64(df4.iloc[:,1]))

# Let's see that the requested Dataframe was obtained.
df4
```

5. Generate a function named as calling_data that take as input the `name of the firm  or list of names (firm_name)` and `(start_year, end_year)`. The function should return a tuple with a dataframe with the same format as the one resulting from task 1-4, and your dictionary of columns. The function should work without `end_year`. It means it only takes `start_year`.

In the following lines of code we implement the required function. The detailed explanation is shown on the same lines.

```python
def call_data(firm_name,start_year,end_year=2023,df4=df4): # By setting 'end_year=2023' we make this value equal to 2023 by default and do not necessarily require an input. 
                                                           # By setting 'df4=df4' we make this value equal to df4 by default and does not necessarily require an input.
    # In the following loop each name is taken from the name list 'firm_name':
    for name in firm_name: 
        
        # With this we verify if the company name exists within the dataset.
        c=(df4['Company'] == name).any() # c is True if the company name exists, False otherwise.
        
        if c==False: 
            # If c is False, then the following is printed and the function terminates.
            result='One of the company names entered is not found within the dataset.'
            print(result)
            return result
        
        else:
            # If c is True, then the following lines continue executing.
            
            # With this we verify if star_year has a value that exists within the years that the DataFrame has for the ORs.
            d=(df4['Year'] == start_year).any() # d is True if the start year exists, False otherwise.
            
            # With this we verify if end_year has a value that exists within the years that the DataFrame has for the ORs.
            d_=(df4['Year'] == end_year).any() # d_ is True if the end year exists, False otherwise.
            
            if d==False or d_==False:
                # If any of the years does not exist in the dataset, then the following is printed and the function terminates.
                result='One of the years entered is not found in the dataset.'
                print(result)
                return result
        
            else:
                # If both years entered exist, then it continues executing.
                
                # e contains a pd.series of booleans where is True if the company name is different from the one entered.
                e=df4['Company']!=name
                # With pd.drop and using e we delete all the rows that contain companies different from the one entered.
                df5=df4.drop(df4[e].index) # df.drop removes rows where the boolean series has True.
                
                # f contains a pd.string of booleans where is True if the year is less than the starting year entered.
                f=df5['Year']<start_year
                # With pd.drop and using f we eliminate all the rows that contain years less than the start year entered.
                df5=df5.drop(df5[f].index)
                
                # g contains a pd.string of booleans where is True if the year is greater than the final year entered.
                g=df5['Year']>end_year
                # With pd.drop and using g we eliminate all the rows that contain years greater than the final year entered.
                df5=df5.drop(df5[g].index)
                
                if firm_name.index(name)==0:
                    # If the company name is the first in the list of names entered then the following Dataframe is created.
                    df6=df5.copy(deep=True) # The created Dataframe is a copy of the previously obtained dataframe.
                else:
                    # If the company name is not the first in the list of names entered, then we concatenate it to the first Dataframe created from df5.copy
                    df6=pd.concat([df6,df5],ignore_index=True) # with ignore_index=True we reset the index every time we concatenate.
                
                # We create the required tuple, the first element will be the Dataframe with the indicated format, and the second element is the previously created dictionary.
                result=(df6,dic)
    
    return result
```

6. Generate a class that do the same as the function, but has a method named as `descripitive` that gives you the mean, max, min, median, and rolling average.

We implement the required class in the following lines of code. The detailed explanation is shown on the same lines.

```python
class Data_Caller:
    
    # We initialize the objects of the class with __init__:
    def __init__(self, df=df4):
        
        # With this we make the class by default have as its initial attribute the dataset that contains all the company names and all the OR years.
        self.df = df 
        
    def call_data(self, firm_name, start_year, end_year=2023): # We create the call_data method, which in structure will be the same as the previously created function.
        
        df4 = self.df # We obtain the DataFrame initially entered.
        
        # What follows has a similar structure to the function created above and already explained in detail above. Therefore we skip the explanation of the following lines.
        for name in firm_name:
            c = (df4['Company'] == name).any()
            
            if c == False:
                result = 'One of the company names entered is not found within the data set.'
                print(result)
                return result
                
            else:
                d=(df4['Year'] == start_year).any() 
                d_=(df4['Year'] == end_year).any() 
            
                if d==False or d_==False:
                    result= 'One of the years entered is not found in the dataset.'
                    print(result)
                    return result
                
                else:
                    e = df4['Company'] != name
                    df5 = df4.drop(df4[e].index)
                    
                    f = df5['Year'] < start_year
                    df5 = df5.drop(df5[f].index)
                    
                    g = df5['Year'] > end_year
                    df5 = df5.drop(df5[g].index)
                    
                    if firm_name.index(name) == 0:
                        df6 = df5.copy(deep=True)
                    else:
                        df6 = pd.concat([df6, df5], ignore_index=True)
                    
                    # Now, the difference is that we will create the following attributes.
                    self.df_result=df6 # We save the resulting dataframe in the 'df_result' attribute
                    self.dic=dic # We save the dictionary created in the 'dic' attribute.
                    self.start=start_year # We save the start year of the data in the 'start' attribute 
                    self.end=end_year # We save the final year of the data in the 'end' attribute
                    
                    result = (self.df_result, self.dic) # We create the required tuple using the created attributes.
                    
        return result
    
    def descriptive(self): # We additionally create this method that requires the question.
        
        # First we get the resulting DataFrame that we got earlier using the attribute we created for it.
        h = self.df_result
        # Then we get the descriptive statistics of this Dataframe.
        h_ = h.describe().drop(columns='Year') # We drop the column since it is not of interest to see the descriptive statistics on this variable.
        
        # Now, to look at the descriptive statistics at the level of each company, we first group the Dataframe data according to the company with the groupby method.
        i = h.groupby([h.columns[0]])
        # Then we get the descriptive statistics of this Dataframe.
        i_ = i.describe().drop(columns='Year') # We drop the column since it is not of interest to see the descriptive statistics on this variable.
        
        # Now, to add the rolling average to the descriptive statistics we do the following:
        if self.end-self.start>2:
            # If the resulting dataset has more than two years of OR recorded, then we will find the rolling average considering that each window will have the last 3 years.
            j=i.rolling(window=3).mean()
            
        elif self.end-self.start==2:
            # If the resulting dataset has only two years of OR recorded then we will find the rolling average considering that each window will have the last 2 years.
            j=i.rolling(window=2).mean()
            
        else: 
            # If the resulting dataset has only one year of OR recorded then we will not find the rolling average and will simply print the following message:
            j='The rolling average is not found as there are less than two years of ORs recorded in the resulting data set.'
            print(j)
            return h,i_,j
        
        return h_,i_,j
```

Now we check if the created class has the same functionality as the `call_data` function.

First we create an object using this class:

```python
Companies=Data_Caller()
```

Now we use the `call_data` method using `firm_name` and `start_year` as inputs which were created previously when the `call_data` function was tested.

```python
Companies.call_data(firm_name,start_year)
```

You can see that the same tuple is obtained as was obtained using the `call_data` function. Now we use the `descriptive` method.

```python
d_1,d_2,d_3=Companies.descriptive()
```

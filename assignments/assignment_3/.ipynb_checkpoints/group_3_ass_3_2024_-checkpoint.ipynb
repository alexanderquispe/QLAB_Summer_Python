## RESOLUCIÓN 3.2
(1) Primero eliminaremos las columnas que tengan sus ingresos operativos en dólares (simplemente usando drop con aquellas columnas que contengan USD) (2) Ahora conservamos las columnas desde "Operating revenue (Turnover) th LCU 2023" hasta el final, junto con el nombre de la empresa. Aquí usaremos "melt" para remodelar nuestro DataFrame (df) de ancho a larg, tal y como se solicita en primer lugar en el ejercicio. (3) Por último, remodelamos el conjunto de datos de formato ancho a largo a nivel de año con "melt" y "stack", el cual nos permite reorganizar la estructura del DataFrame para tener una representación más flexible, donde colocaremos a las variables asociadas a los ingresos operativos en una sola columna llamada 'Año' y sus valores correspondientes en otra columna llamada 'Total'. Esto nos facilitará el análisis y la manipulación de los datos en un formato más estructurado y manejable.

```python
#(1)
columns_to_drop = [col for col in df.columns if 'USD' in col]
df.drop(columns=columns_to_drop, inplace=True)
column_names = df.columns.tolist()
print("Nombres de las columnas:", column_names)
```

```python
#(2)
selected_columns = ['Company name Latin alphabet'] + [col for col in df.columns if 'Operating revenue (Turnover)\nth LCU' in col]
df2 = df[selected_columns]
df2
```

```python
#(3)
#Utilizando la forma de stack
df2_stack = df2.set_index(  ['Company name Latin alphabet']  ).stack().reset_index().rename( {"level_1" : "Año", 
                                                                                0 : "Total" }, axis = 1 )
df2_stack.head()
```

```python
#(3)
#Utilizando melt
df_melt = df2.melt( 
                    id_vars = ['Company name Latin alphabet'] ,
                    var_name = 'Año', 
                    value_name = 'Total'
                 )
df_melt
```

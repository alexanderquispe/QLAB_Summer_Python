### Exercise 5

```python
#First, we create a function where a boolean is defined by the finding, or not, of '.edu'
def edu(emails):
    if emails.find(".edu") == -1:
        print(False)
    if emails.find(".edu") > -1:
        print(True)
```

```python
#Then we create a list with the booleans
edu = list(map(edu,emails))

#We print the list
print(edu)
```

```python
#Then we create a function to find the '@' in each string of the list with booleans
at_list = list(map(lambda item:item[item.find('@')]=='@', new_list))
print(at_list)
```

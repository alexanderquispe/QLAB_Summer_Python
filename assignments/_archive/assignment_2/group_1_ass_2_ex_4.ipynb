```python
import numpy as np
```

```python
#1
#lista de números pares
#Forma 1
##Explicación: np.arrange funciona como un tipo de generador de números de acuerdo a los parámetros que se le indiquen, de inicio (start), final (stop) y los saltos (step)
## aunque inicia con el número indicado no incluye el número que se indica como final

even_numbers = (np.arange(start= 2,stop= 2001, step= 2))
even_numbers

#Forma 2
##Explicación: En este caso, por default si no se indica el step se avanza de 1 en 1, por eso se indica posteriormente que la división de los números
##generados debe dar 0, para solo optener números pares
lista_1 = np.arange(2, 2001)
lista_1 = lista_1[lista_1%2==0]
lista_1
```

```python
#Lista de números impares
#Forma 1
##Explicación: En este caso, se indica por separado (como objetos seperados) en inicio y el final del conjunto de números 

start= 901
end= 1900
odd_numbers = (np.arange(start, end, 2))
odd_numbers

#Forma 2
##Explicación: Aquí se usa un loop, como no está permitido solo lo ponemos a modo de demostración

for j in range(900, 1900):
  if j % 2 != 0:
    print(j)
    
```

```python
##Si revisamos el tipo de objeto que hemos generado, se trata de numpy.darray, sin embargo, se nos pide que sean listas
type(even_numbers)
type(odd_numbers)
```

```python
##Convertimos los objetos antes mencionados en listas
even_numbers = even_numbers.tolist()
odd_numbers = odd_numbers.tolist() 
```

```python
##Ahora revisamos el tipo de objeto, se trata de una lista
type(even_numbers)
type(odd_numbers)
```

```python
#Generamos un objeto que junte ambas listas
numbers = zip(even_numbers, odd_numbers)
numbers 
```

```python
#Como vemos está en formato zip
type(numbers)
```

```python
#Ahora almacenamos el conjunto de números en una lista
list(numbers)
```

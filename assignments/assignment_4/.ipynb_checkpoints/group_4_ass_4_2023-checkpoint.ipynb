1. You work at a college admissions office. When inspecting a dataset of college applicants, you notice that some students have represented their grades with letters (`"AD"`, `"A"`, `"B"`, `"C"`), whereas others have represented their grades with a number between 0 and 20.

    You realize that for consistency, all of the grades should be formatted in the same way, and you decide to format them all as letters. For the conversion, you decide to assign:

    - `"AD"` - any grade 18-20, inclusive
    - `"A"` - any grade 14-17, inclusive
    - `"B"` - any grade 10-13, inclusive
    - `"C"` - any grade <10

    Write a function get_grade() that takes as input:
    - `score` - an integer 0-20 corresponding to a numerical grade

   It should return a Python string with the letter grade that it corresponds to. For instance,

    - A score of 12 corresponds to a B grade. In other words, `get_grade(12)` should return `"B"`.
    - A score of 3 corresponds to an C grade. In other words, `get_grade(3)` should return `"C"`.

   Make sure that when supplying the grade that is returned by the function, it is enclosed in quotes. (For instance, if you want to return `"A"`, you should write `return "A"` and not `return A`.)

```python
def get_grade(score): #We define the function get grade
    if score >= 18: #If the score is equal or over 18, the grade is AD
        return "AD"
    elif score >= 14: #We add another condition for the scores between 14 and 17
        return "A"
    elif score >= 10: #We add one last condition for the scores between 10 and 13
        return "B"
    else:
        return "C" #We ask the function to return the letter C for any other score
```

```python
#Now we can try the function for different scores
get_grade(7) #We see that it returns a C when we have 7 as a score
```

2. You are a programmer at a water agency. Recently, you have been tasked to write a function `get_water_bill()` that takes as input:

    - `num_gallons` = the number of gallons of water that a customer used that month. (This will always be an integer with no decimal part.)

   It should output the water bill.

   The water agency uses this pricing structure:

| Tier        | Amount in gallons | Price per 1000 gallons |
| ----------- | ----------------  | ---------------------- |
| Tier 1      | 0 - 8,000         | 5                      |
| Tier 2      | 8,001 - 22,000    | 6                      |
| Tier 3      | 22,001 - 30,000	  | 7                      |
| Tier 4      | 30,001+	      	  | 10                     |


```python
#First we define our fucntion using the if methos
def get_water_bill(num_gallons:int):#In the definition we sugest that num_gallons must be a int
    if 0<num_gallons <= 8000:#For tier 1 we include all numbersfrom 0 tu 8000
        #divide the gallons by 1000 and multiply it by the price per 1000 gallons from the tier 
        Tier1 = num_gallons*5/1000 #we save it in a new object
        return print("The water bill this month is",Tier1) #we return the phrase with the answer
    #We repeat the process for each tier according to the especifications
    elif 8000<num_gallons<=22000:
        Tier2 =num_gallons*6/1000
        return print("The water bill this month is",Tier2)
    elif 22000<num_gallons<=30000: 
        Tier3 =num_gallons*7/1000
        return print("The water bill this month is",Tier3)
    else:
        Tier4=num_gallons*10/1000
        return print("The water bill this month is",Tier4)
```

```python
#we try our function for every parameter so we can make sure every tier is well made
get_water_bill(8000)
```

## 1.2 While Loop


1. How a game will stay active as long as the player has enough power? Make a variable called `power`, and set its initial value to 5. So, the player's power starts out at 5 and is allowed to keep playing as long as their power is over 0. 

    - Inside the while loop, print a message that reports the player's current power.
    - Inside the while loop, write a statement that decreases the player's power.
    - Outside the while loop, print a message reporting that the player has grown too weak, and that the game is over.




```python
power = 5 #We set the initial value of power to 5
while power > 0:
    print("Player current power is", power) #We ask the algoritm to print the current power
    power -= 1 #We set a cutline when the power is lower than 1

print("Player has grown too weak. Game over.") #We print the message out of the loop when the game is over
```
